```python
import torch

PROJECT_ROOT = Path.cwd().parent
if PROJECT_ROOT not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from src.dataset import build_amino_acid_vocab
from src.model import ProteinClassifier


vocab = build_amino_acid_vocab()

model = ProteinClassifier(
    vocab_size=len(vocab),
    embedding_dim=64,
    hidden_dim=128,
    output_dim=4,
)

sample_sequences = torch.tensor(
    [
        [12, 10, 18, 2, 0],
        [12, 5, 7, 0, 0],
    ],
    dtype=torch.long,
)

hidden_state = model(sample_sequences)

print("Input shape:")
print(sample_sequences.shape)

print("\nFinal hidden-state shape:")
print(hidden_state.shape)

print("\nFinal hidden state for first sequence:")
print(hidden_state[0, 0])

print(model)
```
